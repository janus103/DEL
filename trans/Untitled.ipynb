```python
import os
import sys
import numpy as np
import csv
import dwt as DD
from PIL import Image
```

```python
# # Check that the correct number of arguments are provided
# if len(sys.argv) != 3:
#     print('Usage: python program.py folder_path results.csv')
#     sys.exit()

# Get the folder path and results file name from the command-line arguments
folder_path = '/home/oem/servers/datasets/classification/cifar10_Contrast/val/airplane'#sys.argv[1]
results_file = 'Contrast.csv'#sys.argv[2]

# Load all image files from the folder
image_files = [f for f in os.listdir(folder_path) if f.endswith('.png') or f.endswith('.jpg')]

# Apply dwt() function to each image and calculate the average of absolute values and overall average
# Save results to CSV file
with open(results_file, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['File Name','LL_LL','LL_LH','LL_HL','LL_HH','LH_LL','LH_LH','LH_HL','LH_HH','HL_LL','HL_LH','HL_HL','HL_HH','HH_LL','HH_LH','HH_HL','HH_HH',])
    
    results1 = list()
    
    for idx, file in enumerate(image_files):
        #print(os.path.join(folder_path, file))
        img = Image.open(os.path.join(folder_path, file))
        img = img.convert('L')
        # Convert PIL image to NumPy array
        img = np.array(img)
        print('before ', img.shape)
        
        #img = np.expand_dims(img, axis=-1)
        print(img.shape)
        #img = np.load(os.path.join(folder_path, file))
        LL, LH, HL, HH = DD.get_freqs_1(img)
        LL_LL, LL_LH, LL_HL, LL_HH = DD.get_freqs_1(LL)
        LH_LL, LH_LH, LH_HL, LH_HH = DD.get_freqs_1(LH)
        HL_LL, HL_LH, HL_HL, HL_HH = DD.get_freqs_1(HL)
        HH_LL, HH_LH, HH_HL, HH_HH = DD.get_freqs_1(HH)
        results = [LL_LL, LL_LH, LL_HL, LL_HH, \
                   LH_LL, LH_LH, LH_HL, LH_HH, \
                   HL_LL, HL_LH, HL_HL, HL_HH, \
                   HH_LL, HH_LH, HH_HL, HH_HH
        ]
        temp_lst = list()
        temp_lst.append(file)
        for idx,item in enumerate(results):
            #print('->',(item.shape)) # (8, 8)
            temp_lst.append(np.mean(item))
            #temp_lst.append(np.var(item))
        
        results1.append(temp_lst)
        
        #print(temp_lst)
        if idx % 100 == 0:
            print(idx)
        
    writer.writerows(results1)
       
```
